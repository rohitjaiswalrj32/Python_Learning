# 1. Introduction to OOP

## What is OOP?

* Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects". Objects are instances of classes that encapsulate data (attributes) and behavior (methods) within a single unit. OOP allows for the creation of reusable and modular code, making it easier to manage and maintain.

* Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around data, or objects, rather than functions and logic. An object can be defined as a data field that has unique attributes and behavior.

* In OOP, objects are instances of classes, where a class is a blueprint for creating objects. Classes define the properties (attributes) and behaviors (methods) that the objects created from them can have.

## Why OOP?

* Modularity: OOP allows breaking down a program into smaller, manageable, and interchangeable modules or classes, improving code readability and maintainability.
* Reusability: Through inheritance, classes can inherit attributes and methods from other classes, promoting code reuse.
* Real-World Modeling: Provides a natural way to model real-world entities and their interactions.
* Maintainability: OOP promotes code maintainability, making it easier to modify, extend, and debug large-scale projects.
* Collaboration: OOP can facilitate collaboration among developers, as it provides a structured way to organize and manage code.
* Ease of Debugging: Since code is organized into discrete units (classes), debugging becomes more straightforward. Problems can be isolated to specific classes, reducing the scope of the issue and making it easier to identify and fix bugs.
* Separation of Concerns: OOP promotes the separation of concerns by dividing a program into distinct classes and objects, each responsible for a specific part of the functionality. This modular approach makes code easier to manage, understand, and maintain.
* Interchangeability: Because polymorphic classes can be substituted for one another, developers can design more flexible and reusable code. For example, a function that processes a collection of Animal objects can work with Dog, Cat, or Bird objects without modification.
* Layered Security: OOP allows for the implementation of security layers within objects and classes, making it easier to enforce security policies and protect sensitive data.

# Core Concepts of OOP

## Classes and Objects

* **Class:** A class is a blueprint for creating objects. It defines a set of attributes (variables) and methods (functions) that the objects created from the class will have. Classes encapsulate data and behavior, providing a structure for the objects.

* **Object:** An object is an instance of a class. When a class is defined, no memory is allocated until an object of that class is created. Each object can have different attributes (states) but shares the same methods (behaviors) as other objects of the same class.

### Creating a Class in Python

A class in Python is created using the class keyword, followed by the class name and a colon. The class name should typically follow the PascalCase naming convention (e.g., MyClass).

```python
class Car:
    # Constructor method to initialize the object
    def __init__(self, make, model, year):
        self.make = make       # Instance attribute
        self.model = model     # Instance attribute
        self.year = year       # Instance attribute

    # Method to start the car's engine
    def start_engine(self):
        return f"{self.make} {self.model} engine started!"
```

* __init__ Method: The __init__ method is a special method in Python, known as a constructor. It is automatically called when a new instance of the class is created. It initializes the object's attributes with the values passed as arguments.

* Instance Attributes: Attributes like make, model, and year are instance attributes. They are specific to each object created from the class.

* Methods: Methods like start_engine define the behavior of the objects created from the class.

### Creating and Using Objects

Once a class is defined, you can create objects (instances of the class) by calling the class name as if it were a function, passing any required arguments.

```python
# Creating an object of the Car class
my_car = Car("Toyota", "Corolla", 2020)

# Accessing the object's attributes
print(f"My car is a {my_car.year} {my_car.make} {my_car.model}.")  # Output: My car is a 2020 Toyota Corolla.

# Calling a method on the object
print(my_car.start_engine())  # Output: Toyota Corolla engine started!
```

* Object Creation: my_car = Car("Toyota", "Corolla", 2020) creates a new object of the Car class and initializes its attributes with the given values.
* Accessing Attributes: my_car.year, my_car.make, and my_car.model access the attributes of the my_car object.
* Calling Methods: my_car.start_engine() calls the start_engine method on the my_car object, which returns a string indicating that the car's engine has started.

## Attributes and Methods

### Attributes in Python

Attributes are variables that store data related to an object. In a class, attributes are defined in two main ways:

* Instance Attributes: These are attributes that belong to an instance (object) of a class. They are defined within the __init__ method or any other method using self.

* Class Attributes: These are attributes that are shared across all instances of a class. They are defined directly in the class body, outside of any methods.


**Example: Instance and Class Attributes**

```python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis lupus familiaris"

    def __init__(self, name, age):
        # Instance attributes (unique to each instance)
        self.name = name
        self.age = age

# Creating instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Bella", 5)

# Accessing instance attributes
print(dog1.name)  # Output: Buddy
print(dog2.age)   # Output: 5

# Accessing class attribute
print(dog1.species)  # Output: Canis lupus familiaris
print(dog2.species)  # Output: Canis lupus familiaris
```

* Instance Attributes: name and age are instance attributes. Each Dog object (e.g., dog1, dog2) has its own values for these attributes.
* Class Attribute: species is a class attribute. It is shared by all instances of the Dog class, so both dog1 and dog2 have the same species.

###  Methods in Python

Methods are functions defined inside a class that describe the behaviors of the objects. There are several types of methods in Python:

* Instance Methods: The most common type of methods. They operate on instances of the class and can access and modify the objects attributes. They require self as the first parameter.
* Class Methods: These methods operate on the class itself, rather than on instances of the class. They take cls (which refers to the class) as their first parameter and are defined using the @classmethod decorator.
* Static Methods: These methods do not operate on instances or the class itself. They behave like regular functions but belong to the classs namespace. They are defined using the @staticmethod decorator.

```python
class MathOperations:
    # Class attribute
    pi = 3.14159

    def __init__(self, value):
        # Instance attribute
        self.value = value

    # Instance method
    def multiply_by_pi(self):
        return self.value * MathOperations.pi

    # Class method
    @classmethod
    def circle_area(cls, radius):
        return cls.pi * (radius ** 2)

    # Static method
    @staticmethod
    def add(x, y):
        return x + y

# Using instance methods
operation = MathOperations(5)
print(operation.multiply_by_pi())  # Output: 15.70795

# Using class methods
print(MathOperations.circle_area(3))  # Output: 28.27431

# Using static methods
print(MathOperations.add(10, 20))  # Output: 30
```

* Instance Method (multiply_by_pi): Operates on the instance of MathOperations (operation in this case). It uses the self parameter to access the instances attribute (value) and the class attribute (pi).
* Class Method (circle_area): Operates on the class itself, not on an instance. It uses the cls parameter to access the class attribute pi and calculates the area of a circle.
* Static Method (add): Acts like a regular function but is part of the class. It doesnt access or modify class or instance attributes.

###  Introduction to Inheritance

Inheritance is one of the fundamental concepts of Object-Oriented Programming (OOP). It allows a new class (called the child or derived class) to inherit attributes and methods from an existing class (called the parent or base class). This promotes code reusability and establishes a natural hierarchy between classes.

**Benefits of Inheritance**

* Code Reusability: Inheritance allows a new class to reuse the code of an existing class, avoiding duplication and promoting the DRY (Don't Repeat Yourself) principle.
* Extensibility: You can extend the functionality of an existing class by adding new methods and attributes to the derived class without modifying the base class.
* Hierarchy Representation: Inheritance models real-world hierarchies and relationships between different entities, making the code more intuitive and easier to understand.

### Types of Inheritance in Python

**Single Inheritance**
In single inheritance, a child class inherits from a single parent class.

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

# Child class
class Dog(Animal):
    def speak(self):
        return f"{self.name} barks."

# Creating an object of the Dog class
dog = Dog("Buddy")
print(dog.speak())  # Output: Buddy barks.
```

**Explanation:**
* Parent Class (Animal): Defines a generic animal with a name attribute and a speak method.
* Child Class (Dog): Inherits from Animal and overrides the speak method to provide specific behavior for a dog.

**Multiple Inheritance**
In multiple inheritance, a child class can inherit attributes and methods from more than one parent class.

```python
# Parent class 1
class Flyer:
    def fly(self):
        return "Flying high!"

# Parent class 2
class Swimmer:
    def swim(self):
        return "Swimming in the water!"

# Child class
class Duck(Flyer, Swimmer):
    def quack(self):
        return "Quack quack!"

# Creating an object of the Duck class
duck = Duck()
print(duck.fly())   # Output: Flying high!
print(duck.swim())  # Output: Swimming in the water!
print(duck.quack()) # Output: Quack quack!
```

**Explanation:**
* Parent Class 1 (Flyer): Defines a fly method.
*Parent Class 2 (Swimmer): Defines a swim method.
* Child Class (Duck): Inherits both fly and swim methods from Flyer and Swimmer, respectively, and adds its own quack method.


**Multilevel Inheritance**
In multilevel inheritance, a class is derived from a class that is also derived from another class.

```python
# Base class
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        return f"{self.brand} vehicle starting."

# Intermediate class
class Car(Vehicle):
    def start(self):
        return f"{self.brand} car starting."

# Derived class
class ElectricCar(Car):
    def start(self):
        return f"{self.brand} electric car starting silently."

# Creating an object of the ElectricCar class
tesla = ElectricCar("Tesla")
print(tesla.start())  # Output: Tesla electric car starting silently.
```

**Explanation:**
* Base Class (Vehicle): The root class with a start method.
* Intermediate Class (Car): Inherits from Vehicle and overrides the start method.
* Derived Class (ElectricCar): Inherits from Car and further overrides the start method to reflect the specific behavior of electric cars.

**Hierarchical Inheritance**
In hierarchical inheritance, multiple classes inherit from a single parent class.

```python
# Parent class
class Shape:
    def area(self):
        pass  # Abstract method, should be implemented in derived classes

# Child class 1
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * (self.radius ** 2)

# Child class 2
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Creating objects of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle area: {circle.area()}")        # Output: Circle area: 78.53975
print(f"Rectangle area: {rectangle.area()}")  # Output: Rectangle area: 24
```

**Explanation:**
* Parent Class (Shape): Acts as a blueprint for different shapes, defining a method area that must be implemented by derived classes.
* Child Classes (Circle and Rectangle): Inherit from Shape and provide their own implementations of the area method.

### Access Modifiers in Python

Python uses a system of prefixes to indicate the level of access control for class attributes and methods:

* Public: Attributes and methods that are accessible from anywhere. There is no special prefix for public members (they are defined normally).
* Protected: Attributes and methods that should not be accessed directly outside the class or its subclasses. In Python, protected members are prefixed with a single underscore (_).
* Private: Attributes and methods that are not accessible outside the class. They are intended to be completely hidden from outside access. In Python, private members are prefixed with a double underscore (__).


**Example: Public, Protected, and Private Members**

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name              # Public attribute
        self._salary = salary          # Protected attribute
        self.__id = 1001               # Private attribute

    def display_employee_info(self):
        return f"Name: {self.name}, Salary: {self._salary}, ID: {self.__id}"

    def __get_id(self):                # Private method
        return self.__id

# Creating an object of the Employee class
emp = Employee("John Doe", 50000)

# Accessing public attribute
print(emp.name)  # Output: John Doe

# Accessing protected attribute (not recommended)
print(emp._salary)  # Output: 50000

# Accessing private attribute (will raise an AttributeError)
try:
    print(emp.__id)  # This will cause an error
except AttributeError as e:
    print(e)  # Output: 'Employee' object has no attribute '__id'

# Accessing private attribute using name mangling
print(emp._Employee__id)  # Output: 1001

# Accessing private method using name mangling
print(emp._Employee__get_id())  # Output: 1001
```

**Explanation:**
* Public Attribute (name): Accessible from outside the class using emp.name.
* Protected Attribute (_salary): Intended to be accessible only within the class and its subclasses, but can still be accessed outside the class using emp._salary. However, this is not recommended.
* Private Attribute (__id): Cannot be accessed directly from outside the class. Attempting to do so results in an AttributeError. However, it can be accessed using name mangling (emp._Employee__id).


###  Benefits of Encapsulation



* Control of Data Access: By using private and protected members, encapsulation allows you to control who can access or modify specific attributes, ensuring that the internal state of an object is maintained consistently.
* Data Hiding: Encapsulation hides the internal implementation details of a class from the outside world, providing a clear interface for interaction.
* Improved Security: Encapsulation can prevent accidental changes to the internal state of an object, thereby improving the security of the code.

### Introduction to Polymorphism

* Polymorphism is one of the core concepts of Object-Oriented Programming (OOP). The term "polymorphism" means "many shapes" or "many forms." In programming, polymorphism refers to the ability of different objects to respond to the same method in their own unique way. It allows one interface to be used for a general class of actions, where the specific action is determined by the exact nature of the situation.
* Polymorphism in Python allows objects of different classes to be treated as objects of a common superclass. It provides the flexibility to design cleaner, more maintainable, and more modular code. Through method overriding, method overloading, and duck typing, polymorphism enables objects to respond to the same interface in different ways, which is a cornerstone of OOP and a powerful tool for developers.

### Ways to achieve polymorphism

* Method Overriding: A child class provides a specific implementation of a method that is already defined in its parent class.
* Method Overloading: Although Python does not support traditional method overloading (same method name with different parameters), it can be achieved through default arguments or variable-length arguments.
* Duck Typing: A concept where the type of an object is determined by its behavior (methods it implements) rather than its inheritance.


#### Method Overriding



Method overriding allows a subclass to provide a specific implementation of a method that is already defined in its parent class. This is the most common way to achieve polymorphism in Python.

```python
class Animal:
    def speak(self):
        return "Animal sound"

class Dog(Animal):
    def speak(self):
        return "Bark"

class Cat(Animal):
    def speak(self):
        return "Meow"

# Creating objects of Dog and Cat
dog = Dog()
cat = Cat()

# Calling the speak method on different objects
print(dog.speak())  # Output: Bark
print(cat.speak())  # Output: Meow
```

* Parent Class (Animal): Defines a generic speak method.
* Child Classes (Dog and Cat): Override the speak method to provide specific sounds for dogs and cats.
* In this example, the speak method is polymorphic because it behaves differently depending on the object that calls it.

#### Method Overloading in Python

While Python does not support traditional method overloading, you can achieve similar functionality using default arguments or *args and **kwargs.

```python
class MathOperations:
    def add(self, a, b=0, c=0):
        return a + b + c

# Creating an object of MathOperations
math_ops = MathOperations()

# Calling the add method with different numbers of arguments
print(math_ops.add(5))        # Output: 5
print(math_ops.add(5, 10))    # Output: 15
print(math_ops.add(5, 10, 15)) # Output: 30
```

Method (add): Can accept one, two, or three arguments, thanks to default values for b and c. This is a form of method overloading.


#### Duck Typing

Duck typing is a concept related to polymorphism where the type or class of an object is less important than the methods it defines. In duck typing, an object's suitability is determined by the presence of certain methods and properties, rather than the actual type of the object.

The phrase "If it looks like a duck, swims like a duck, and quacks like a duck, then it probably is a duck" is often used to describe duck typing.

```python
class Dog:
    def speak(self):
        return "Bark"

class Cat:
    def speak(self):
        return "Meow"

class Bird:
    def speak(self):
        return "Chirp"

# Function demonstrating duck typing
def animal_sound(animal):
    print(animal.speak())

# Creating objects of Dog, Cat, and Bird
dog = Dog()
cat = Cat()
bird = Bird()

# Passing different objects to the same function
animal_sound(dog)   # Output: Bark
animal_sound(cat)   # Output: Meow
animal_sound(bird)  # Output: Chirp
```

* Function (animal_sound): Accepts any object that has a speak method, regardless of the object's class. This is a demonstration of duck typing in action.
* In this example, as long as an object has a speak method, it can be passed to the animal_sound function, and the function will work correctly.

### Polymorphism with Functions and Objects

Polymorphism can also be used in functions and objects to achieve flexibility in code. Different classes can be passed to the same function, and the function will execute methods defined in those classes.

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Function demonstrating polymorphism
def print_area(shape):
    print(f"Area: {shape.area()}")

# Creating objects of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Passing different objects to the same function
print_area(circle)    # Output: Area: 78.53975
print_area(rectangle) # Output: Area: 24
```

**Explanation:** Function (print_area): Can accept any object that is a subclass of Shape and has an area method. This allows the same function to be used with different shapes.

## Abstraction

### Introduction to Abstraction

* Abstraction is one of the four fundamental concepts in Object-Oriented Programming (OOP), along with encapsulation, inheritance, and polymorphism. Abstraction refers to the process of hiding the implementation details of a particular functionality and exposing only the essential features to the user. In simpler terms, abstraction allows you to focus on what an object does rather than how it does it.

* In Python, abstraction is achieved using abstract classes and interfaces, which provide a blueprint for other classes to implement. An abstract class can have one or more abstract methods, which are methods that are declared but contain no implementation.

###  Abstract Classes and Methods

In Python, an abstract class is a class that cannot be instantiated on its own. It is intended to be subclassed, and its abstract methods must be implemented by its subclasses.

Python provides the abc (Abstract Base Classes) module to implement abstraction. This module provides a way to define abstract classes and methods using the ABC class and the @abstractmethod decorator.

**Example: Abstract Class and Abstract Method**

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass  # Abstract method, no implementation

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

# Trying to instantiate an abstract class will raise an error
try:
    animal = Animal()
except TypeError as e:
    print(e)  # Output: Can't instantiate abstract class Animal with abstract method sound

# Creating objects of Dog and Cat, which are subclasses of Animal
dog = Dog()
cat = Cat()

print(dog.sound())  # Output: Bark
print(cat.sound())  # Output: Meow
```

* Abstract Class (Animal): Cannot be instantiated because it contains an abstract method (sound).
* Abstract Method (sound): Declared in the Animal class without any implementation.
* Concrete Classes (Dog and Cat): Subclasses of Animal that provide specific implementations of the sound method.
* In this example, the Animal class defines an abstract method sound, which is then implemented by its subclasses Dog and Cat. Attempting to create an instance of Animal directly results in an error, enforcing that the abstract method must be implemented in the subclass.

### The Role of Abstraction

* Simplification: Abstraction allows complex systems to be broken down into simpler, more manageable components. By focusing on high-level operations, developers can ignore the intricate details of implementation.
* Code Reusability: Abstract classes can serve as a blueprint for multiple subclasses. This encourages code reuse and standardization across different parts of an application.
* Maintainability: Since abstract classes define a clear contract for what methods a subclass must implement, they make code easier to understand and maintain. Changes in the abstract class can be propagated across all subclasses, ensuring consistency.

###  Interfaces in Python


While Python does not have a built-in concept of interfaces like some other programming languages (e.g., Java), the concept can be mimicked using abstract classes. An interface is essentially an abstract class with only abstract methodsmethods that need to be implemented by any concrete class that inherits the interface.

**Example: Mimicking Interfaces with Abstract Classes**

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * (self.radius ** 2)

    def perimeter(self):
        return 2 * 3.14159 * self.radius

# Creating objects of Rectangle and Circle
rectangle = Rectangle(4, 5)
circle = Circle(3)

print(f"Rectangle Area: {rectangle.area()}")  # Output: Rectangle Area: 20
print(f"Rectangle Perimeter: {rectangle.perimeter()}")  # Output: Rectangle Perimeter: 18

print(f"Circle Area: {circle.area()}")  # Output: Circle Area: 28.27331
print(f"Circle Perimeter: {circle.perimeter()}")  # Output: Circle Perimeter: 18.84954
```

* Interface (Shape): Acts as a blueprint that requires any subclass to implement the area and perimeter methods.
* Concrete Classes (Rectangle and Circle): Implement the required methods from the Shape interface, providing specific functionality.
* In this example, the Shape abstract class defines the contract for any shape to implement area and perimeter methods.Rectangle and Circle classes provide concrete implementations of these methods.
